```python
from dandi.dandiapi import DandiAPIClient

dandiset_id = '000041'  # ephys dataset from the Svoboda Lab
filepath = 'sub-BWRat17/sub-BWRat17_ses-BWRat17-121712_ecephys.nwb'  # 1.5GB file
with DandiAPIClient() as client:
    asset = client.get_dandiset(dandiset_id, 'draft').get_asset_by_path(filepath)
    s3_url = asset.get_content_url(follow_redirects=1, strip_query=True)
```

```python
dandiset_id, filepath="0000041","sub-BWRat17/sub-BWRat17_ses-BWRat17-121712_ecephys.nwb"
from pynwb import NWBHDF5IO
from nwbwidgets import nwb2widget

io = NWBHDF5IO(s3_url,mode='r',load_namespaces=True,driver='ros3')
nwb = io.read()
```

```python
nwb2widget(nwb)
```

```python
nwb.units.to_dataframe()
```

```python
import numpy
spiketrainN=[]
outputspiketrainN=[]
spiketrain=numpy.zeros((len(nwb.units["spike_times"]),100))
outputspiketrain=numpy.zeros((len(nwb.units["spike_times"]),100))


for i in range(len(nwb.units["spike_times"])):
    #print(i)
    spiketrain[i,:] = numpy.histogram(nwb.units["spike_times"][i], 
                         bins = 100, 
                         range = (0, 1))[0]/0.1
    
    outputspiketrain[i,:] = numpy.histogram(nwb.units["spike_times"][i], 
                         bins = 100, 
                         range = (0.1, 1.1))[0]/0.1


# select only excitatory neurons & firing rate not equal to zero   

totalspikerate=numpy.sum(spiketrain,axis=1)+numpy.sum(outputspiketrain,axis=1)

spiketrainN=spiketrain[(nwb.units["cell_type"][:]=="excitatory") & (totalspikerate>0),:]
outputspiketrainN=outputspiketrain[(nwb.units["cell_type"][:]=="excitatory") & (totalspikerate>0),:]

  
```

```python
from pyglmnet import GLMCV
from pyglmnet.datasets import fetch_group_lasso_datasets
import matplotlib.pyplot as plt
glm = GLMCV(distr="poisson", tol=1e-3,
            score_metric="pseudo_R2",
            alpha=1.0, learning_rate=3, max_iter=100,cv=3, verbose=0)
```

```python
spiketrainN[1]
```

```python
Beta=numpy.zeros((len(outputspiketrainN[:,0]),len(outputspiketrainN[:,0])))

for i in range(len(outputspiketrainN[:,0])):
    glm.fit(spiketrainN.transpose(), outputspiketrainN[i,:])
    Beta[i,:]=glm.beta_;
    
```
